# 1. Sistema de Recomendação de Filmes

Sistemas de recomendação são ferramentas de informação que ajudam os usuários a encontrar os itens que desejam a partir do grande número de itens disponíveis. O principal objetivo do sistema de recomendação é prever a classificação que um usuário específico atribui a um item. Ajuda o usuário a encontrar a melhor solução na lista de itens disponíveis. [1]

Muitas empresas utilizam sistemas de recomendação para atender seu usuário e aumentar seus lucros como Netflix, YouTube, Amazon e outros. Hoje em dia o que compramos online é recomendação. Por exemplo, se quisermos comprar livros, ouvir música, assistir filmes, etc., existe um sistema de recomendação que funciona em segundo plano e sugere o usuário com base em suas ações anteriores. [1]

A recomendação de filmes é a aplicação mais usada por sistemas de recomendação, juntamente com plataformas de multimídia online, que visa ajudar os clientes a acessar os filmes preferidos de forma inteligente a partir de uma enorme biblioteca de filmes. [2]

Existem basicamente três tipos de sistemas de recomendação:

- Recomendador Demográfico: Oferece recomendações generalizadas para cada usuário, com base na popularidade do filme. Este sistema recomenda os mesmos filmes com características demográficas semelhantes a todos os usuários.

- Recomendador baseado em conteúdo: Ele constrói um mecanismo que calcula a similaridade entre filmes com base em certas métricas (como gênero, diretor, descrição, atores) e sugere filmes que são mais semelhantes a um determinado filme que o usuário gostou. A ideia geral por trás desses sistemas de recomendação é que se uma pessoa gostou de um item específico, ela também gostará de um item semelhante a ele.

- Recomendador Colaborativo: Este sistema combina pessoas com interesses semelhantes e fornece recomendações com base nessa correspondência.


Neste notebook, usamos o conjunto de dados MovieLens e implementamos três algoritmos de recomendação, incluindo os 3 recomendadores citados acima.  

Os conjuntos de dados MovieLens incluem:

- "tmdb_5000_credits" que contém as seguintes features:
    * movie_id - um identificador exclusivo para cada filme.
    * title - Título de cada filme.
    * cast - o nome dos atores principais e coadjuvantes.
    * crew - o nome do diretor, editor, compositor, escritor, etc. <br><br>

- "tmdb_5000_movies" que contém os seguintes recursos:
    * budget - o orçamento no qual o filme foi feito.
    * genre - o gênero do filme, ação, comédia, suspense etc.
    * homepage - Um link para a homepage do filme.
    * id - Este é o movie_id como no primeiro conjunto de dados.
    * keywords - as palavras-chave ou tags relacionadas ao filme.
    * original_language - O idioma no qual o filme foi feito.
    * original_title - O título do filme antes da tradução ou adaptação.
    * overview - uma breve descrição do filme.
    * popularity - uma quantidade numérica que especifica a popularidade do filme.
    * production_companies - a produtora do filme.
    * production_countries - o país em que foi produzido.
    * release_date - a data em que foi lançado.
    * revenue - a receita mundial gerada pelo filme.
    * runtime - o tempo de execução do filme em minutos.
    * status - "Liberado" ou "Rumor".
    * slogan - slogan do filme.
    * title - Título do filme.
    * vote_average - avaliações médias que o filme recebeu.
    * vote_count - a contagem dos votos recebidos.<br><br>

Os outros arquivos csv correspondem aos mesmos de cima, mas com particionamentos diferentes:

- "ratings_small" que contém as seguintes features:
    * userId - um identificador exclusivo para cada usuário.
    * movieId - um identificador exclusivo para cada filme.
    * classification - A classificação dada pelo usuário para cada filme.
    * timestamp - a hora em que a classificação foi registrada. <br><br>

- movies_metadata.csv: o arquivo principal dos metadados dos filmes. Contém informações sobre 45.000 filmes apresentados no conjunto de dados Full MovieLens. As features incluem 'posters', 'backdrops', 'budget', 'revenue', 'release dates', 'languages', 'production countries' e 'companies'.<br><br>

- keywords.csv: Contém as palavras-chave do enredo do filme para nossos filmes MovieLens. Disponível na forma de um objeto JSON textificado.<br><br>

- Credit.csv: Consiste em informações sobre o elenco e a equipe de todos os nossos filmes. Disponível na forma de um objeto JSON textificado.<br><br>

O conjunto de dados Full MovieLens, que consiste em 26 milhões de classificações e 750.000 aplicações de tag de 270.000 usuários em todos os 45.000 filmes neste conjunto de dados, pode ser acessado [aqui](https://grouplens.org/datasets/movielens/latest/).

### 1.1.2. Outras Análises

```python
def load_tmdb_credits(path):
    df = pd.read_csv(path)
    json_columns = ['cast', 'crew']
    for column in json_columns:
        df[column] = df[column].apply(json.loads)
    return df

def load_tmdb_movies(path):
    df = pd.read_csv(path)
    df['release_date'] = pd.to_datetime(df['release_date']).apply(lambda x: x.date())
    json_columns = ['genres', 'keywords', 'production_countries', 'production_companies', 
                    'spoken_languages']
    for column in json_columns:
        df[column] = df[column].apply(json.loads)
    return df

credits = load_tmdb_credits("tmdb_5000_credits.csv")
movies = load_tmdb_movies("tmdb_5000_movies.csv")
```

```python
#Para consistência entre os dois dataframes, vamos renomear o campo 'id' no dataframe do filme 
#como 'movie_id'. 
movies.rename(columns={"id": "movie_id"}, inplace=True)
```

#### 1.1.2.1 Limpeza de Dados

Podemos ver que alguns dos dados em certos campos são muito brutos nos conjuntos de dados de créditos e filmes. A maioria deles contém muitas informações de que não precisamos, então vamos limpá-los um pouco. Primeiro, vamos começar com o conjunto de dados de créditos e ver uma amostra dos campos de elenco e equipe de créditos. 

```python
credits["cast"].iloc[0]
```

```python
credits["crew"].iloc[0]
```

Seguindo a metodologia descrita [aqui](https://www.kaggle.com/sohier/tmdb-format-introduction), podemos substituir o dataframe de créditos por um dataframe de pessoas que exibe o papel desempenhado por atores e membros da equipe em vários filmes em um muito mais limpa e gerenciável. 

```python
credits.apply(lambda row: [x.update({'movie_id': row['movie_id']}) for x in row['cast']], axis=1)
credits.apply(lambda row: [x.update({'movie_id': row['movie_id']}) for x in row['crew']], axis=1)
credits.apply(lambda row: [person.update({'order': order}) for order, person in enumerate(row['crew'])], axis=1)

cast_list = []
credits["cast"].apply(lambda x: cast_list.extend(x))
cast = pd.DataFrame(cast_list)
cast["type"] = "cast"

crew_list = []
credits["crew"].apply(lambda x: crew_list.extend(x))
crew = pd.DataFrame(crew_list)
crew["type"] = "crew"

people = pd.concat([cast, crew], ignore_index=True, sort=True)
del credits
```

```python
people.head()
```

Vamos agora passar para o conjunto de dados de filmes. Alguns dos campos dos filmes contêm muitas informações desnecessárias e também podem ser bastante difíceis de ler. Esses campos são as colunas 'genres', 'keywords', 'production_companies', 'production_countries' e 'spoken_languages'. Os dados nesses campos podem ser mais adequados para aparecer em seu próprio dataframe para cada campo, para que possamos discernir melhor quais fatores de um determinado campo aparecem para um determinado filme. Devemos, portanto, criar um dataframe separado para cada um dos campos mencionados acima e removê-los do conjunto de dados de filmes. 

```python
movies.apply(lambda row: [x.update({'movie_id': row['movie_id']}) for x in row['genres']], axis=1)
movies.apply(lambda row: [x.update({'movie_id': row['movie_id']}) for x in row['keywords']], axis=1)
movies.apply(lambda row: [x.update({'movie_id': row['movie_id']}) for x in row['production_companies']], axis=1)
movies.apply(lambda row: [x.update({'movie_id': row['movie_id']}) for x in row['production_countries']], axis=1)
movies.apply(lambda row: [x.update({'movie_id': row['movie_id']}) for x in row['spoken_languages']], axis=1)

genres = []
movies["genres"].apply(lambda x: genres.extend(x))
genres = pd.get_dummies(pd.DataFrame(genres).drop("id", axis=1).set_index("movie_id")).sum(level=0)
genres.rename(columns = lambda x: str(x)[5:], inplace=True)

keywords = []
movies["keywords"].apply(lambda x: keywords.extend(x))
keywords = pd.get_dummies(pd.DataFrame(keywords).drop("id", axis=1).set_index("movie_id")).sum(level=0)
keywords.rename(columns = lambda x: str(x)[5:], inplace=True)

production_companies = []
movies["production_companies"].apply(lambda x: production_companies.extend(x))
production_companies = pd.get_dummies(pd.DataFrame(production_companies).drop("id", axis=1).set_index("movie_id")).sum(level=0)
production_companies.rename(columns = lambda x: str(x)[5:], inplace=True)

production_countries = []
movies["production_countries"].apply(lambda x: production_countries.extend(x))
production_countries = pd.get_dummies(pd.DataFrame(production_countries).drop("iso_3166_1", axis=1).set_index("movie_id")).sum(level=0)
production_countries.rename(columns = lambda x: str(x)[5:], inplace=True)

spoken_languages = []
movies["spoken_languages"].apply(lambda x: spoken_languages.extend(x))
spoken_languages = pd.get_dummies(pd.DataFrame(spoken_languages).drop("iso_639_1", axis=1).set_index("movie_id")).sum(level=0)
spoken_languages.rename(columns = lambda x: str(x)[5:], inplace=True)

movies.drop(["genres", "keywords", "production_companies", "production_countries", "spoken_languages"], axis=1, inplace=True)
```

```python
movies.head()
```

Existem alguns campos nos filmes que provavelmente não são muito úteis para nós. Provavelmente não há muitas informações úteis que possamos obter por ter a página inicial de um filme ou seu título original, portanto, excluiremos esses campos. Como já temos informações sobre os gêneros de um determinado filme e palavras-chave relacionadas a ele, provavelmente não precisamos ter uma overview do filme ou de seu slogan.

```python
movies.drop(["homepage", "original_title", "overview", "tagline"], axis=1, inplace=True)
```

Presumivelmente, o status da maioria dos filmes seria 'Lançado', então provavelmente podemos descartar esse campo também, mas por curiosidade, vamos olhar um gráfico de barras que mostra a frequência dos status dos filmes no conjunto de dados. 

```python
sns.countplot(movies["status"])
```

Claramente, o status dos filmes a maioria é 'Lançado', então simplesmente removeremos esse campo do conjunto de dados. 

```python
movies.drop(["status"], axis=1, inplace=True)
```

O idioma original dos filmes no conjunto de dados também pode ser a maioria em inglês e, portanto, não fornecer muitas informações úteis para nós. Vamos dar uma olhada na distribuição dos principais idiomas originais. 

```python
movies["original_language"].value_counts().sort_values().tail().plot.barh()
```

Como podemos ver, os filmes no conjunto de dados são predominantemente em inglês. Portanto, este campo provavelmente não renderá muitas informações úteis para nós e pode ser descartado. 

```python
movies.drop(["original_language"], axis=1, inplace=True)
```

Quando se trata da data de lançamento de um filme, provavelmente estaríamos menos interessados no dia exato em que foi lançado e mais curiosos para ver em que ano foi lançado ou em que mês do ano foi lançado. É adicionado dois campos ao conjunto de dados dos filmes, um para o ano de lançamento e um para o mês de lançamento, e o campo release_date é removido. 

```python
movies["release_date"] = pd.to_datetime(movies["release_date"])
movies["release_year"] = movies["release_date"].dt.year
movies["release_month"] = movies["release_date"].dt.month
movies.drop(["release_date"], axis=1, inplace=True)
```

```python
#Também definimos o movie_id como o índice do dataframe de nossos filmes 
movies.set_index("movie_id", inplace=True)
```

Como podemos ver, há uma correlação pronunciada entre o orçamento do filme e o rendimento, com filmes com custos de produção mais altos geralmente gerando mais receita. Há também uma correlação positiva interessante, embora menos pronunciada, entre o tempo de execução do filme e o rendimento, com filmes mais longos geralmente acumulando rendimento mais alto. O ano em que um filme foi lançado também não parece ter muito impacto no rendimento que gera, portanto, não podemos necessariamente esperar que os filmes mais recentes gerem mais rendimento do que os filmes mais antigos. Vendo os boxplots comparando o mês de lançamento do filme com o rendimento, podemos ver que a maioria dos filmes de maior bilheteria parecem ter sido lançados nos meses de maio, junho, novembro e dezembro, com os filmes lançados em janeiro, setembro e outubro parecendo ser menos bem-sucedido financeiramente. Quanto mais rendimento um filme acumula, sem surpresa, também parece estar correlacionado com sua pontuação de popularidade e o número de votos que recebe. Filmes de alta bilheteria também parecem obter classificações de votos mais altas, embora essa correlação positiva seja muito menos pronunciada do que entre rendimento e popularidade e contagem de votos.

### 1.2.1. Recomendador Demográfico

O sistema de recomendação demográfica gera recomendações com base nos atributos demográficos do usuário. Ele categoriza os usuários com base em seus atributos e recomenda os filmes utilizando seus dados demográficos. <br><br>


A implementação da Filtragem Demográfica é direta. Tudo o que temos a fazer é classificar nossos filmes com base nas classificações e exibir os filmes principais de nossa lista. Portanto, devemos:

- Criar uma métrica para pontuar ou avaliar os filmes.

- Calcular a pontuação de cada filme.

- Inserir as pontuações e recomendar aos usuários o filme mais bem avaliado.<br><br>


Podemos usar a classificação média do filme como pontuação, mas usar isso não seria muito justo, pois um filme com classificação média de 8.9 e apenas 3 votos não pode ser considerado melhor do que o filme com classificação média de 7.8, mas 40 votos. Então, usamos a classificação ponderada do IMDB.

```python
df_movielens['vote_average'].mean()
```

Portanto, a classificação média para todos os filmes é de aproximadamente 6 em uma escala de 10. 

A próxima etapa é determinar um valor apropriado para m, os votos mínimos necessários para serem listados no gráfico. Usaremos o percentil 90 como nosso ponto de corte. Em outras palavras, para um filme fazer sucesso, ele deve ter mais votos do que pelo menos 90% dos filmes da lista. 

```python
df_movielens['vote_count'].quantile(q=0.9)
```

```python
#Agora, podemos filtrar os filmes que se qualificam para o gráfico 
df_filtered=df_movielens[df_movielens['vote_count']>df_movielens['vote_count'].quantile(q=0.9)]
```

```python
df_filtered.shape
```

Vemos que existem 481 filmes que se qualificam para esta lista. Agora, precisamos calcular nossa métrica para cada filme qualificado. Para fazer isso, definiremos uma função e definiremos uma nova pontuação de feature, da qual calcularemos o valor aplicando esta função ao nosso DataFrame de filmes qualificados: 

```python
def movie_score(x):
    v=x['vote_count']
    m=df_movielens['vote_count'].quantile(q=0.9)
    R=x['vote_average']
    C=df_movielens['vote_average'].mean()
    return ((R*v)/(v+m))+((C*m)/(v+m))
```

```python
df_filtered['score']=df_filtered.apply(movie_score, axis=1)
```

```python
"""
Finalmente, classificamos o dataframe com base no recurso de pontuação. Em seguida, mostramos o título, gêneros, contagem de votos, 
média de votos, popularidade e pontuação dos 10 filmes principais. 
"""
df_highscore=df_filtered.sort_values(by='score', ascending=False).head(10)
```

```python
df_highscore[['title', 'vote_count', 'vote_average', 'popularity', 'score']]
```

```python
df_popular= df_movielens.sort_values('popularity', ascending=False).head(10)
```

### 1.2.2. Recomendador baseado em conteúdo 

Esta filtragem recomenda o item ao usuário com base em sua experiência anterior. Por exemplo, se um usuário gosta apenas de filmes de ação, o sistema prevê para ele apenas filmes de ação semelhantes aos quais ele classificou bem. Ao contrário da filtragem colaborativa, a filtragem baseada em conteúdo não enfrenta novos problemas de usuário. Ele não tem outra interação do usuário nele. Ele lida apenas com o interesse específico do usuário. <br>
A filtragem baseada em conteúdo primeiro verifica a preferência do usuário e, em seguida, sugere-o com os filmes ou qualquer outro produto para ele. Ele se concentra apenas nas ideias e pensamentos de um único usuário e fornece previsões com base em seu interesse. Portanto, se falamos de filmes, a técnica de filtragem baseada em conteúdo verifica a classificação dada pelo usuário. A abordagem verifica quais filmes recebem avaliações altas do usuário, verificando as categorias de gênero no perfil do usuário. Após analisar o perfil do usuário, a técnica recomenda filmes ao usuário de acordo com seu gosto.

Foi construído dois recomendadores baseados em conteúdo, incluindo:

- Visão geral do filme

#### 1.2.2.1.  Recomendador baseado na visão geral do filme

Primeiro, calculamos as pontuações de similaridade de pares para todos os filmes com base em sua coluna 'overview'. Em seguida, recomendamos filmes com base nessa pontuação. 

```python
df_movielens['overview'].head()
```

```python
df_movielens['overview'].isnull().sum()
```

```python
#Substituímos NaN na coluna 'overview' por uma string vazia 
df_movielens['overview'].fillna(' ', inplace=True)
```

```python
#Agora calculamos os vetores Term Frequency-Inverse Document Frequency (TF-IDF) para cada visão geral 
tfidfv=TfidfVectorizer(analyzer='word', stop_words='english')
tfidfv_matrix=tfidfv.fit_transform(df_movielens['overview'])
print(tfidfv_matrix.todense())
tfidfv_matrix.todense().shape
```

Vemos que mais de 20.000 palavras diferentes foram usadas para descrever os 4803 filmes em nosso conjunto de dados. 

```python
#Computando pontuação de similaridade
cosine_sim1 = linear_kernel(tfidfv_matrix, tfidfv_matrix)
cosine_sim1.shape 
```

Agora temos uma matriz de similaridade de cosseno par a par para todos os filmes em nosso conjunto de dados. 

A próxima etapa é definir uma função de recomendação que recebe o título de um filme como uma entrada e produz uma lista dos 10 filmes mais semelhantes. Para fazer isso:

1. Precisamos de um mapeamento reverso de títulos de filmes e índices de dataframe. Em outras palavras, construímos uma série para identificar o índice de um filme em nosso dataframe, de acordo com seu título.

2. A função deve obter o índice do filme dado seu título.

3. Obter a lista de pontuações de similaridade de cosseno para aquele filme específico com todos os filmes. Convertê-lo em uma lista de tuplas em que o primeiro elemento é sua posição e o segundo é a pontuação de similaridade.

4. Classificar a lista de tuplas mencionada anteriormente com base nas pontuações de similaridade; ou seja, o segundo elemento.

5. Obter os 10 principais elementos desta lista. Ignorar o primeiro elemento, pois ele se refere a si mesmo (o filme mais semelhante a um determinado filme é o próprio filme).

6. Retornar os títulos correspondentes aos índices dos elementos principais. 

```python
#Definindo função de recomendação
indices=pd.Series(data=list(df_movielens.index), index= df_movielens['title'] )
```

```python
indices.head()
```

```python
def content_recommendations(title, cosine_sim):
    idx = indices[title]
    
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    sim_scores.sort(key=lambda x: x[1], reverse=True)

    sim_scores=sim_scores[1:11]
    
    ind=[]
    for (x,y) in sim_scores:
        ind.append(x)
        
    tit=[]
    for x in ind:
        tit.append(df_movielens.iloc[x]['title'])
    return pd.Series(data=tit, index=ind)
```

```python
content_recommendations('The Dark Knight Rises',cosine_sim1)
```

```python
content_recommendations('Avatar',cosine_sim1)
```

Embora nosso sistema tenha feito um trabalho decente ao encontrar filmes com visões gerais e descrições semelhantes, a qualidade das recomendações não é tão boa. "The Dark Knight Rises" retorna todos os filmes do Batman, embora seja mais provável que as pessoas que gostaram desse filme estejam mais inclinadas a desfrutar de outros filmes de Christopher Nolan. Isso é algo que não pode ser capturado pelo sistema atual. 

#### 1.2.2.2. Recomendador baseado em gêneros, keywords, elenco, equipe do filme 

Para melhorar a qualidade do recomendador baseado em conteúdo, usamos metadados melhores. Portanto, construímos um recomendador com base nos seguintes metadados:

- o diretor
- os 3 atores principais
- os 3 principais gêneros relacionados
- as 3 principais palavras-chave do enredo do filme

Das features equipe, elenco, gêneros e palavras-chave, precisamos extrair o diretor e os três atores, gêneros e palavras-chave mais importantes associados a esse filme. 

```python
#Preprocessando os conteudos 
type(df_movielens['cast'].iloc[0])
```

```python
features = ['cast', 'crew', 'keywords', 'genres']
for feature in features:
    df_movielens[feature] = df_movielens[feature].apply(literal_eval)
```

```python
type(df_movielens['cast'].iloc[0])
```

```python
# Pegua o nome do diretor pela feature 'crew'. Se o diretor não estiver listado, retorna NaN 
def get_director(x):
    for a in x:
        if a['job']=='Director':
            return a['name'] 
    return 'NaN'
```

```python
## Obtém os 3 principais elementos da lista ou a lista inteira; o que ocorrer mais nas colunas de elenco, gêneros e palavras-chave. 
def get_top3(x):
    new=[]
    for a in x[:3]:
        new.append(a['name']) 
    return new

    return []
```

```python
df_movielens['director']=df_movielens['crew'].apply(lambda x: get_director(x))
```

```python
df_movielens['actor']=df_movielens['cast'].apply(lambda x:get_top3(x))
```

```python
df_movielens['genres']=df_movielens['genres'].apply(lambda x:get_top3(x))
```

```python
df_movielens['keywords']=df_movielens['keywords'].apply(lambda x:get_top3(x))
```

```python
df_movielens[['title', 'actor', 'director', 'keywords', 'genres']].head(3)
```

```python
#A próxima etapa seria converter os nomes e as ocorrências de palavras-chave em letras minúsculas e remover todos os espaços entre eles. Isso é feito para que nosso vetorizador não conte o Johnny de "Johnny Depp" e "Johnny Galecki" como o mesmo 
def clean_director(x):
    return x.lower().replace(' ','')
```

```python
def clean_top3(x):
    new=[]
    for a in x:
        new.append(a.lower().replace(' ',''))
    return new
```

```python
df_movielens['director']=df_movielens['director'].apply(lambda x: clean_director(x))
```

```python
df_movielens['actor']=df_movielens['actor'].apply(lambda x:clean_top3(x))
```

```python
df_movielens['keywords']=df_movielens['keywords'].apply(lambda x:clean_top3(x))
```

```python
df_movielens['genres']=df_movielens['genres'].apply(lambda x:clean_top3(x))
```

```python
df_movielens[['title', 'actor', 'director', 'keywords', 'genres']].head(3)
```

```python
#Agora criamos a coluna 'sopa', que contém todos os metadados que queremos alimentar ao nosso vetorizador (atores, diretor, gêneros e palavras-chave)
def create_soup(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['actor']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])
```

```python
df_movielens['soup'] = df_movielens.apply(create_soup, axis=1)
```

```python
#construindo matrix tf-idf
cv = CountVectorizer(stop_words='english')
cv_matrix = cv.fit_transform(df_movielens['soup'])
```

```python
#Computando pontuação de similaridade
cosine_sim2 = cosine_similarity(cv_matrix, cv_matrix)
```

```python
content_recommendations('Avatar', cosine_sim2)
```

```python
content_recommendations('The Matrix', cosine_sim2)
```

Vemos que nosso recomendador teve sucesso em capturar mais informações devido a mais metadados e nos deu recomendações melhores. 

### 1.2.3. Recomendador colaborativo


A Filtragem Colaborativa (CF) é uma das técnicas mais conhecidas para recomendar filmes. Esta técnica sugere um filme relevante para o usuário com base na escolha do vizinho. Ele primeiro descobre a semelhança entre o usuário e seu vizinho e, em seguida, prevê os itens. Pode haver um número n de usuários. Esta técnica encontra o usuário semelhante na lista de usuários. Mas a semelhança entre os usuários é descoberta com base nas avaliações que os usuários deram a um filme específico. Desta forma, a abordagem continua e o resultado desejado é gerado. Esta estratégia leva classificações dadas pelo usuário para qualquer item do grande catálogo do catálogo de itens de classificações dadas pelo usuário. Este grande catálogo é conhecido como matriz de itens do usuário.

![collab_filt.png](collab_filt.png "Title")

Nosso recomendador baseado em conteúdo sofre de algumas limitações severas. Ele só é capaz de sugerir filmes próximos a um determinado filme. Ou seja, não é capaz de capturar preferências e fornecer recomendações entre gêneros.

Além disso, o mecanismo que construímos não é realmente pessoal, pois não captura as preferências pessoais de um usuário. Qualquer pessoa que consultar nosso mecanismo para recomendações baseadas em um filme receberá as mesmas recomendações para aquele filme, independentemente de quem ele / ela seja.

Portanto, nesta seção, usaremos uma técnica chamada Filtragem Colaborativa para fazer recomendações aos telespectadores. É basicamente de dois tipos: 

- Recomendação baseada em usuário: Esses sistemas recomendam produtos para um usuário que usuários semelhantes gostaram. Para medir a semelhança entre dois usuários, podemos usar a correlação de Pearson ou a semelhança de cosseno. Embora computar o CF baseado no usuário seja muito simples, ele apresenta vários problemas. Um dos principais problemas é que as preferências dos usuários podem mudar com o tempo. Isso indica que o pré-cálculo da matriz com base nos usuários vizinhos pode levar a um desempenho ruim. Para resolver este problema, podemos aplicar o CF baseado em item.
- Recomendação baseada em item: Esta abordagem também é conhecida como filtragem item a item. É usado para recomendar qualquer item com base na classificação do item relacionado. Ao analisar a classificação, os usuários cujas classificações são semelhantes para itens diferentes, apenas aqueles itens estão disponíveis para recomendação. Tem sido amplamente utilizado por todos os gigantes da Web, incluindo Netflix, YouTube etc. 

Neste trabalho, apresentamos o recomendador colaborativo com 3 algoritmos diferentes:
- SVD
- KNN
- Embedding

#### 1.2.3.1 Usando o SVD

Uma maneira de lidar com o problema de escalabilidade e esparsidade criado pelo CF é aproveitar um modelo de fator latente para capturar a similaridade entre usuários e itens. Essencialmente, queremos transformar o problema de recomendação em um problema de otimização. Podemos ver como somos bons em prever a classificação dos itens dados a um usuário. Uma métrica comum é Root Mean Square Error (RMSE). Quanto menor o RMSE, melhor será o desempenho. Como não sabemos a classificação dos itens não vistos, iremos ignorá-los temporariamente. Ou seja, estamos apenas minimizando o RMSE nas entradas conhecidas na matriz de utilidade. Para atingir RMSE mínimo, a Decomposição de Valor Singular (SVD) é adotada. 

A biblioteca Surprise é utilizada para construir e analisar sistemas de recomendação que lidam com dados de classificação explícitos. Aqui, usamos a biblioteca Surprise, que usa algoritmos extremamente poderosos, como Singular Value Decomposition (SVD), para minimizar o Root Mean Square Error (RMSE), que é medido pela Validação Cruzada Kfold e oferece ótimas recomendações.

Para o filme com ID 302, temos uma previsão estimada de 2,63. Uma característica surpreendente desse sistema de recomendação é que ele não se importa com o que o filme é (ou o que ele contém). Ele funciona puramente com base em uma ID de filme atribuída e tenta prever as classificações com base em como os outros usuários previram o filme.

SVD lida com o problema de escalabilidade e dispersão apresentado pelo CF com sucesso. No entanto, SVD não é isento de falhas. A principal desvantagem do SVD é que não há pouca explicação para o motivo pelo qual recomendamos um item a um usuário. Isso pode ser um grande problema se os usuários estiverem ansiosos para saber por que um item específico é recomendado a eles. 

#### 1.2.3.2. Com KNN 

O algoritmo de k-vizinhos mais próximos (KNN) não faz suposições sobre a distribuição de dados subjacente, mas depende da similaridade de recursos do item. Quando um KNN faz uma previsão sobre um filme, ele calcula a “distância” (métricas de distância serão discutidas mais tarde) entre o filme alvo e todos os outros filmes em seu banco de dados. Em seguida, classifica suas distâncias e retorna os k principais filmes vizinhos mais próximos como as recomendações de filmes mais semelhantes.

```python
#Agora vamos gerar uma lista 'genreList' com todos os gêneros únicos possíveis mencionados no conjunto de dados. 
genreList = []
for index, row in movies.iterrows():
    genres = row["genres"]
    
    for genre in genres:
        if genre not in genreList:
            genreList.append(genre)
genreList[:10]
```

```python
#One Hot Encoding para múltiplos labels
def binary(genre_list):
    binaryList = []
    
    for genre in genreList:
        if genre in genre_list:
            binaryList.append(1)
        else:
            binaryList.append(0)
    
    return binaryList
```

```python
words_list = []
for index, row in movies.iterrows():
    genres = row["keywords"]
    
    for genre in genres:
        if genre not in words_list:
            words_list.append(genre)
```

```python
def binary(words):
    binaryList = []
    for genre in words_list:
        if genre in words:
            binaryList.append(1)
        else:
            binaryList.append(0)
    return binaryList
```

```python
movies['words_bin'] = movies['keywords'].apply(lambda x: binary(x))
movies = movies[(movies['vote_average']!=0)] #removing the movies with 0 score and without drector names 
movies = movies[movies['director']!='']
```

```python
from scipy import spatial

#similaridade entre filmes usando cosseno
def Similarity(movieId1, movieId2):
    a = movies.iloc[movieId1]
    b = movies.iloc[movieId2]
    
    genresA = a['genres_bin']
    genresB = b['genres_bin']
    
    genreDistance = spatial.distance.cosine(genresA, genresB)
    
    scoreA = a['cast_bin']
    scoreB = b['cast_bin']
    scoreDistance = spatial.distance.cosine(scoreA, scoreB)
    
    directA = a['director_bin']
    directB = b['director_bin']
    directDistance = spatial.distance.cosine(directA, directB)
    
    wordsA = a['words_bin']
    wordsB = b['words_bin']
    wordsDistance = spatial.distance.cosine(directA, directB)
    return genreDistance + directDistance + scoreDistance + wordsDistance
```

Vemos que a distância é cerca de 1.98, o que é alto. Quanto maior a distância, menos semelhantes os filmes são. Vamos ver o que esses filmes aleatórios realmente eram.

```python
import operator

#preditor de pontuações
def predict_score(name):
    new_movie = movies[movies['original_title'].str.contains(name)].iloc[0].to_frame().T
    print('Selected Movie: ',new_movie.original_title.values[0])
    def getNeighbors(baseMovie, K):
        distances = []
    
        for index, movie in movies.iterrows():
            if movie['new_id'] != baseMovie['new_id'].values[0]:
                dist = Similarity(baseMovie['new_id'].values[0], movie['new_id'])
                distances.append((movie['new_id'], dist))
    
        distances.sort(key=operator.itemgetter(1))
        neighbors = []
    
        for x in range(K):
            neighbors.append(distances[x])
        return neighbors

    K = 10
    avgRating = 0
    neighbors = getNeighbors(new_movie, K)
    
    print('\nRecommended Movies: \n')
    for neighbor in neighbors:
        avgRating = avgRating+movies.iloc[neighbor[0]][2]  
        print( movies.iloc[neighbor[0]][0]+" | Genres: "+str(movies.iloc[neighbor[0]][1]).strip('[]').replace(' ','')+" | Rating: "+str(movies.iloc[neighbor[0]][2]))
    
    print('\n')
    avgRating = avgRating/K
    print('The predicted rating for %s is: %f' %(new_movie['original_title'].values[0],avgRating))
    print('The actual rating for %s is %f' %(new_movie['original_title'].values[0],new_movie['vote_average']))
```

#### 1.2.3.3. Com Embedding

Muitos algoritmos de Filtragem Colaborativa são baseados. Recentemente, vários trabalhos na área de Processamento de Linguagem Naturasugeriram aprender uma representação latente de palavras usando algoritmos neurais embedding. 

```python
import string
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

from tensorflow.keras import models
from tensorflow.keras import layers
from tensorflow.keras import optimizers
from tensorflow.keras import losses
from tensorflow.keras import regularizers
from tensorflow.keras import metrics
from tensorflow.keras.utils import plot_model
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.text import text_to_word_sequence
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
metadata = pd.read_csv('the movies dataset/movies_metadata.csv')
```

```python
credits = pd.read_csv('the movies dataset/credits.csv')
keywords = pd.read_csv('the movies dataset/keywords.csv')

metadata = metadata.drop([19730, 29503, 35587])

# convertendo ids para int, necessário para merging
keywords['id'] = keywords['id'].astype('int')
credits['id'] = credits['id'].astype('int')
metadata['id'] = metadata['id'].astype('int')

# Merge keywords e credits no dataframe principal
metadata = metadata.merge(credits, on='id')
metadata = metadata.merge(keywords, on='id')
```

```python
# limitar os dados para nn estourar a memoria
metadata = metadata[:20000]
```

```python
metadata = metadata.reset_index()
```

```python
ratings = pd.read_csv('the movies dataset/ratings_small.csv')
ratings.head()
```

```python
movies = metadata[['id', 'title', 'genres']]
movies['genres'] =  ["".join(st) for st in movies['genres']] 
movies.head()
```

```python
pd.crosstab(ratings['userId'], movies['id']).head()
```

```python
encoder = LabelEncoder()
ratings['user'] = encoder.fit_transform(ratings['userId'].values)
n_users = ratings['user'].nunique()

ratings['movie'] = encoder.fit_transform(ratings['movieId'].values)
n_movies = ratings['movie'].nunique()


ratings['rating'] = ratings['rating'].values.astype(np.float32)

min_rating = min(ratings['rating'])
max_rating = max(ratings['rating'])

n_users, n_movies, min_rating, max_rating
```

```python
X = ratings[['user', 'movie']].values
y = ratings['rating'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

Aqui fazemos uma regularização para ajudar a generalização do modelo, através do hiperparâmetro embeddings_regularizer. Essa regularização é feita em todas as tabelas embedding para evitar overfitting.

```python
# Numero de fatores latentes
emb_sz = 50

# User embeddings
user = layers.Input(shape=(1,))
user_emb = layers.Embedding(n_users, emb_sz, embeddings_regularizer=regularizers.l2(1e-6))(user)
user_emb = layers.Reshape((emb_sz,))(user_emb)

# Movie embeddings
movie = layers.Input(shape=(1,))
movie_emb = layers.Embedding(n_movies, emb_sz, embeddings_regularizer=regularizers.l2(1e-6))(movie)
movie_emb = layers.Reshape((emb_sz,))(movie_emb)

# Dot product
rating = layers.Dot(axes=1)([user_emb, movie_emb])
rating = layers.Activation('sigmoid')(rating)
rating = layers.Lambda(lambda x:x*(max_rating - min_rating) + min_rating)(rating) # scalling up

# Modelo
model = models.Model([user, movie], rating)

# Compilando o modelo
model.compile(loss='mse', metrics=[metrics.RootMeanSquaredError()],
              optimizer=optimizers.Adam(lr=0.001))

# mostrando sumário do modelo
model.summary()
plot_model(model)
```

```python
model.fit(x=[X_train[:,0], X_train[:,1]], y=y_train,
          batch_size=64, epochs=5, verbose=1,
          validation_data=([X_test[:,0], X_test[:,1]], y_test))
```

No código acima, os fatores latentes dependem do produto escalar dos usuários e da classificação do filme. Mas pode haver algum filme anterior ou fator do usuário, que não depende da relação entre o usuário e as classificações altas, por exemplo, quanto o filme é popular, independentemente da avaliação específica do usuário. Digamos que um usuário nunca clica em uma determinada categoria de filmes (populares), como podemos recomendar a ele? Ou o quanto o usuário gosta de filmes, independentemente de sua classificação em um filme específico?

Faremos duas novas tabelas de embeddings para user_bias e movie_bias, e não iremos trabalhar com a camada Dot, mas adicionando após Dot.

Aqui também faça a regularização nas camadas de embeddings user_bias e movie_bias. 

```python
# Numero de fatores latentes
emb_sz = 50

# User embeddings
user = layers.Input(shape=(1,))
user_emb = layers.Embedding(n_users, emb_sz, embeddings_regularizer=regularizers.l2(1e-6))(user)
user_emb = layers.Reshape((emb_sz,))(user_emb)

# User bias
user_bias = layers.Embedding(n_users, 1, embeddings_regularizer=regularizers.l2(1e-6))(user)
user_bias = layers.Reshape((1,))(user_bias)

# Movie embeddings
movie = layers.Input(shape=(1,))
movie_emb = layers.Embedding(n_movies, emb_sz, embeddings_regularizer=regularizers.l2(1e-6))(movie)
movie_emb = layers.Reshape((emb_sz,))(movie_emb)

# Movie bias
movie_bias = layers.Embedding(n_movies, 1, embeddings_regularizer=regularizers.l2(1e-6))(movie)
movie_bias = layers.Reshape((1,))(movie_bias)

# Dot product
rating = layers.Concatenate()([user_emb, movie_emb])

# Adicionando biases
rating = layers.Add()([rating, user_bias, movie_bias])
rating = layers.Dense(10, activation='relu')(rating)
rating = layers.Dense(1, activation='sigmoid')(rating)
rating = layers.Lambda(lambda x:x*(max_rating - min_rating) + min_rating)(rating)

# Modelo
model = models.Model([user, movie], rating)

# Compilando o modelo
model.compile(loss='mse',  metrics=[metrics.RootMeanSquaredError()],
              optimizer=optimizers.Adam(lr=0.001))

# mostrar sumário do modelo
print(model.summary())
plot_model(model, to_file='model.png')
```

```python
model.fit(x=[X_train[:,0], X_train[:,1]], y=y_train,
          batch_size=64, epochs=5, verbose=1,
          validation_data=([X_test[:,0], X_test[:,1]], y_test))
```

Dessa vez o RMSE é aproximadamente 90%.
